```python
sns.scatterplot(x='age', y='charges', data=df, hue='smoker')
plt.show()

sns.scatterplot(x='bmi', y='charges', data=df, hue='smoker')
plt.show()
```

```python
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Charges vs Smoker Status")
plt.show()
```

```python
df.groupby('children')['charges'].mean().sort_index()
```

```python
sns.barplot(x='children', y='charges', data=df, estimator='mean', ci=None)
plt.title("Average Charges by Number of Children")
plt.show()
```

```python
# Average charges by region
print(df.groupby('region')['charges'].mean())

# Boxplot
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='region', y='charges', data=df)
plt.title("Charges by Region")
plt.show()
```

```python
# Average charges by sex
print(df.groupby('sex')['charges'].mean())

# Boxplot
sns.boxplot(x='sex', y='charges', data=df)
plt.title("Charges by Sex")
plt.show()
```

**Insights**

1. Age vs Charges: Charges increase with age
2. Smoker vs Charges: Strongest factor affecting charges
3. BMI vs Charges:
    i. People with higher BMI (obese range >30) tend to have higher charges.
   ii. Among smokers with high BMI, charges increase drastically
4. Children vs Charges: Number of children has little to no effect on charge.
5. Region vs Charges: No significant regional variation.
6. Sex vs Charges: No major difference between male and female in charges.
7. Smoker status, age, and BMI are the key drivers of medical charges.
